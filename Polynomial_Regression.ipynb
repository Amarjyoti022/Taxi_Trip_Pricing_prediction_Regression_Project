```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_excel("Taxi_cleaned.xlsx")
df
```

```python
X = df.drop("Trip_Price",axis = 1)
y = df["Trip_Price"]
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=54)
```

### Modelling

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_converter = PolynomialFeatures(degree=2,include_bias=False)
X_train_poly = pd.DataFrame(polynomial_converter.fit_transform(X_train))
X_test_poly = pd.DataFrame(polynomial_converter.transform(X_test))
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train_poly,y_train)

print("Intercept",model1.intercept_)
print("Coefficient",model1.coef_)
```

### Prediction & Evaluation

```python
ypred_train = model1.predict(X_train_poly)

from sklearn.metrics import mean_squared_error,r2_score
print("Train r2:",r2_score(y_train,ypred_train))

#Prediction on test data
ypred_test = model1.predict(X_test_poly)
print("Test R2:",r2_score(y_test,ypred_test))

from sklearn.model_selection import cross_val_score
print("Cross validation score:",cross_val_score(model1,X_train_poly,y_train,cv=5,scoring="r2").mean())
```
