```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_excel("Taxi_cleaned.xlsx")
df
```

```python
X = df.drop("Trip_Price",axis = 1)
y = df["Trip_Price"]
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=54)
```

**Applying hyperparameter tunning for Ridge regression**

```python
from sklearn.model_selection import GridSearchCV

#model
from sklearn.linear_model import Ridge
estimator = Ridge()

#parameters and values
param_grid = {"alpha":list(range(1,100))}

#Identifying the best value of the parameter within given values for the given data
model_hp = GridSearchCV(estimator,param_grid,cv=5,scoring="r2")

model_hp.fit(X_train,y_train)

model_hp.best_params_
```

### Modelling

```python
from sklearn.linear_model import Ridge
ridge_best = Ridge(alpha=12)
ridge_best.fit(X_train,y_train)

print("Intercept",ridge_best.intercept_)
print("Coefficient",ridge_best.coef_)
```

### Prediction & Evaluation

```python
#Prediction
train_prediction = ridge_best.predict(X_train)
test_prediction = ridge_best.predict(X_test)

#Evaluation
print("Train r2:",ridge_best.score(X_train,y_train))
print("Test R2:",ridge_best.score(X_test,y_test))

from sklearn.model_selection import cross_val_score
print("Cross validation score:",cross_val_score(ridge_best,X,y,cv=5).mean())
```
